### `products` Data Frame

We will begin by addressing the missing values in the `products` data frame.

````python
# Display rows where the product_name column has missing values
# Every row where product_name = Not a Number (NaN)
print(products[products['product_name'].isna()])
````

It seems suspicious that all rows with missing `product_name` might be associated with `aisle_id` 100 and `department_id` 21. Letâ€™s verify this by checking if any rows with missing `product_name` have an `aisle_id` different from 100 and 21. This will help us confirm if the issue is isolated to this aisle or spread across others.

````python
# Combine conditions to check for missing product names in aisles other than 100

print(len(products[
    products['product_name'].isna()
    & (products['aisle_id'] != 100)
    ])
)

# We use len() to return a quantitative result.
# A row count of 0 implies that for all rows where product_name == NaN -> aisle_id == 100
````

````python
# Combine conditions to check for missing product names in departments other than 21

print(len(products[
    products['product_name'].isna()
    & (products['department_id'] != 21)
    ])
)

# A row count of 0 implies that for all rows where product_name == NaN -> department_id == 21
# Therefore, we find that if department_id == 21 && aisle_id == 100, -> product_name == NaN
````

To better understand the missing product_name values, letâ€™s determine what department_id 21 and aisle_id 100 represent by referencing the departments and aisles tables.

````python
# What is this aisle and department?

# We print the departments dataframe where df['department_id'] == 21
print(departments[
    departments['department_id'] == 21
    ]
)
# We print the aisles dataframe where df['aisles_id'] == 100
print(aisles[
    aisles['aisle_id'] == 100
    ]
)

# This shows us the recorded names for the troublesome IDs in the respective tables.
````

````python
# Fill missing product names with 'Unknown' using .fillna() on the product_name column
products['product_name'] = products['product_name'].fillna('Unknown')

# Print where product_name == our new 'Unknown' sentinel value
# note that the row count == the previous NaN row count; 
# aisle_id and department_id vals are all 100 and 21, respectively
print(products[products['product_name'] == 'Unknown'])
````
